```python
from collections import defaultdict

#stores variable number of premises
premises = []
print('Valid operators are "ifthen", "dot", "trip", "not","wedge"')
print('')
prenum = int(input('How many premises? '))
if prenum>0:
    for i in range(prenum):
        premises.append(input(str('Premise #'+str(i+1)+': ')))
else:
    print('Nice try, asshole')


#defines functions for operators (ifthen, trip, not, wedge, dot)
def ifthen(A,B):
    if A==True and B==True:
        return True
    elif A==False:
        return True
    else:
        return False     
def equiv(A,B):
    if A==B:
        return True
    else:
        return False
def opp(A):
    A=not A
def wedge(A,B):
    if A==False and B==False:
        return False
    else:
        return True
def dot(A,B):
    if A==True and B==True:
        return True
    else:
        return False

atoms = []
#builds complete atom list from all premises
for premise in premises:
    for i in range(len(premise)):
        if premise[i].isupper()==True:
            #if a new atom is recognized
            if premise[i] not in atoms:
                atoms.append(premise[i])
            #if a duplicate atom is detected, it is added to trans with its number attached (to give it a unique dict key)
            else:
                pass

rownum = 2**len(atoms)
            
#generates T/F table
TF = defaultdict(list)
for i in atoms:
    TF['%s' % i] = []
    x = int(2**(atoms.index(i)))
    for a in range(int(rownum/(2*x))):
        for b in range(x):
            TF['%s' % i].append(True)
        for b in range(x):
            TF['%s' % i].append(False)
            
#printtable for atoms
patoms = []
ptable = defaultdict(list)
for atom in atoms:
    patoms.append(str(atom+' '))
for i in patoms:
    ptable['%s' % i] = []
    x = int(2**(patoms.index(i)))
    for a in range(int(rownum/(2*x))):
        for b in range(x):
            ptable['%s' % i].append('T')
        for b in range(x):
            ptable['%s' % i].append('F')

#prints table
for row,col in ptable.items():
    print(row,col)
    
#builds PREM for each premise, prints
for premise in premises:
    trans = []

    for i in range(len(premise)):
    #builds atoms list of atomic propositions comprised of capital romans
        if premise[i].isupper()==True:
            trans.append(premise[i])
    #adds first letter of operator to trans
    #checks first two symbols for operators, which must be correct for operator to make it to trans
        elif str(premise[i])=='i'and premise[i+1]=='f':
            trans.append(premise[i])
        elif premise[i]=='t' and premise[i+1]=='r':
            trans.append(premise[i])
        elif premise[i]=='n' and premise[i+1]=='o':
            trans.append(premise[i])
        elif premise[i]=='w' and premise[i+1]=='e':
            trans.append(premise[i])
        elif premise[i]=='d' and premise[i+1]=='o':
            trans.append(premise[i])
        else:
            pass

    #adds '/' to the beginning of each premise
    trans=['/ ']+trans

    #adds instance number to operators in trans to give them a unique dict key
    #this should be lumped in with the 
    for i in range(len(trans)):
        if trans[i]=='/ ':
            pass
        else:
            check = trans[0:i+1]
            checknum = check.count(trans[i])
            trans[i]=str(trans[i]+str(checknum))

    #TF will be dictionary of unique atomic propositions printed before premises
    #This builds out all combos of T/F for atoms stored in dict with atom as key and list of booleans as values
    TF = defaultdict(list)
    for i in atoms:
        TF['%s' % i] = []
        x = int(2**(atoms.index(i)))
        for a in range(int(rownum/(2*x))):
            for b in range(x):
                TF['%s' % i].append(True)
            for b in range(x):
                TF['%s' % i].append(False)


    #This section builds a defaultdict for the premise as transcribed in the trans list
    #operators start out with 'opr' in every entry space
    PREM = defaultdict(list)
    for item in trans:
        PREM['%s' % item] = []
        if item[0].isupper()==True:
            PREM['%s' % item]=TF['%s' % item[0]]
        elif item[0]=='/':
            for a in range(int(rownum)):
                PREM['%s' % item].append('/')
        else:
            for a in range(int(rownum)):
                PREM['%s' % item].append('opr')

    #makes operators work
    #(not super efficient, could use some work)
    #I could probably expand this to include the check and checknum procedure
    for item in trans:
        x = trans.index(item)
        if item[0].islower()==True:   
            for row in range(rownum):
                if item[0] == 'd':
                    PREM['%s' % item][row]=dot(PREM['%s' % trans[x-1]][row],PREM['%s' % trans[x+1]][row])
                elif item[0] == 'i':
                    PREM['%s' % item][row]=ifthen(PREM['%s' % trans[x-1]][row],PREM['%s' % trans[x+1]][row])
                elif item[0] == 'n':
                    PREM['%s' % item][row]=opp(PREM['%s' % trans[x-1]][row])
                elif item[0] == 't':
                    PREM['%s' % item][row]=equiv(PREM['%s' % trans[x-1]][row],PREM['%s' % trans[x+1]][row])
                elif item[0] == 'w':
                    PREM['%s' % item][row]=wedge(PREM['%s' % trans[x-1]][row],PREM['%s' % trans[x+1]][row])
                else:
                    pass
    
    for item in trans:
        for row in range(rownum):
            if PREM['%s' % item][row]==True:
                PREM['%s' % item][row]='T'
            elif PREM['%s' % item][row]==False:
                PREM['%s' % item][row]='F'    
            else:
                pass
    
    #prints table of boolean values in PREM
    for row,col in PREM.items():
        print(row,col)
#Currently there is no error handling for mistyped symbols or lowercase romans;
#if noncap entry doesn't have first two letters in an operator it is passed
```
